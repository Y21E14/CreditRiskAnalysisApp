# Data Cleansing 

```python
#imports
import numpy as np
import pandas as pd 

#Step 1: CSV Reader Node - load and read data - the file has just edited header names
rating_df = pd.read_csv('Set B Corporate Rating_editing.csv')
rating_df
```

```python
#Step 2: Missing Value Node - drop rows that consist of missing values
#checks for missing values in dataset, isnull() method returns DF of the same shape as rating_, where each element is a boolean value indicating whether the corresponding element in ri is missing or not.
rating_df.isnull().sum()
```

```python
#Step 2: Missing Value Node - drop rows that consist of missing values
rating_df = rating_df.dropna()
rating_df
```

```python
#Step 3: Column Filter Node - remove unnecessary columns (axis=1 removes cols) 
rating_df = rating_df.drop(
    #include market value
    ['Dividends per Share - Pay Date - Calendar', 
     'Book Value Per Share', 
     'Common Equity - Liquidation Value', 
     'Comprehensive Income - Parent', 
     'Employees', 
     'Earnings Per Share from Operations'], 
    axis=1, 
    errors='ignore'  # Prevent errors if columns are already missing
)

rating_df
```

```python
#Step 4: Rule Engine Node - new 'rating level' column (3 categories)

# Add a new column rating level and create a funtion for it to split into 3 categories 
def classify_risk(rating):
    if rating in ["CCC-", "CCC", "CCC+", "B-", "CC", "B", "B+"]:
        return "High Risk"
    elif rating in ["BB-", "BB", "BB+", "BBB-", "BBB", "BBB+"]:
        return "Moderate Risk"
    elif rating in ["A-", "A", "A+", "AA-", "AA", "AA+", "AAA"]:
        return "Low Risk"
    else:
        # Default outcome if the value doesn't match any rule
        return None   

# Apply the function to create a new column for rating level
rating_df['Rating Level'] = rating_df['Rating'].apply(classify_risk)

# Display the DataFrame
rating_df



```

```python
#Step 5: Math Formula Node - append new 'total current liabilities' column 
# to remove any leading or trailing whitespace from the column names
rating_df.columns = rating_df.columns.str.strip()

# Check if required columns exist
if 'Total Long-Term Debt' in rating_df.columns and 'Total Debt in Current Liabilities' in rating_df.columns:
    # Perform the calculation and create the new column
    rating_df['Total Current Liabilities'] = rating_df['Total Long-Term Debt'] - rating_df['Total Debt in Current Liabilities']
    print("Successfully added 'Total Current Liabilities'.")
    
else:
    print("Required columns 'Total Long-Term Debt' or 'Total Debt in Current Liabilities' are missing.")

rating_df



#Step 6: Math Formula Node - append new 'debt-to-equity ratio' column
# Check if required columns exist
if 'Total Debt in Current Liabilities' in rating_df.columns and 'Total Stockholders Equity' in rating_df.columns:
    # Perform the calculation and create the new column
    rating_df['Debt-to-Equity Ratio'] = rating_df['Total Debt in Current Liabilities'] / rating_df['Total Stockholders Equity']
    print("Successfully added 'Debt-to-Equity Ratio'.")
    
else:
    print("Required columns 'Total Debt in Current Liabilities' or 'Total Stockholders Equity' are missing.")

rating_df


#Step 7: Math Formula Node - append new 'non-current asset' column 
# Check if required columns exist
if 'Total Asset' in rating_df.columns and 'Cash' in rating_df.columns and 'Total Inventories' in rating_df.columns:
    # Perform the calculation and create the new column
    rating_df['Non current asset'] = rating_df['Total Asset'] - (rating_df['Cash'] + rating_df['Total Inventories'])

    print("Successfully added 'Non current asset'.")
    
else:
    print("Required columns 'Total Asset' or 'Cash' or 'Total Inventories' are missing.")

rating_df


#Step 8: Math Formula Node - append new 'current asset' column 
# Check if required columns exist
if 'Total Asset' in rating_df.columns and 'Non current asset' in rating_df.columns:
    # Perform the calculation and create the new column
    rating_df['Total Current Asset'] = rating_df['Total Asset'] - rating_df['Non current asset']
    print("Successfully added 'Total Current Asset'.")
    
else:
    print("Required columns 'Total Asset' or 'Non current asset' are missing.")

rating_df


#Step 9: Math Formula Node - append new 'working capital ratio' column 
# Check if required columns exist
if 'Total Current Asset' in rating_df.columns and 'Total Current Liabilities' in rating_df.columns:
    # Perform the calculation and create the new column
    rating_df['Working Capital Ratio'] = rating_df['Total Current Asset'] / rating_df['Total Current Liabilities']
    print("Successfully added 'Working Capital Ratio'.")
    
else:
    print("Required columns 'Total Current Asset' or 'Total Current Liabilities' are missing.")

rating_df


#Step 10: Math Formula Node - append new 'debt service coverage ratio' column 
# Check if required columns exist
if 'Earnings Before Interest' in rating_df.columns and 'Total Long-Term Debt' in rating_df.columns and 'Total Debt in Current Liabilities' in rating_df.columns:
    # Perform the calculation and create the new column
    rating_df['Debt Service Coverage Ratio'] = rating_df['Earnings Before Interest'] / (rating_df['Total Long-Term Debt'] + rating_df['Total Debt in Current Liabilities'])
    print("Successfully added 'Debt Service Coverage Ratio'.")
    
else:
    print("Required columns 'Earnings Before Interest' or 'Total Long-Term Debt' or 'Total Debt in Current Liabilities' are missing.")

rating_df


#Step 11: Math Formula Node - append new 'gross profit margin' column 
# Check if required columns exist
if 'Gross Profit (Loss)' in rating_df.columns and 'Total Revenue' in rating_df.columns:
    # Perform the calculation and create the new column
    rating_df['Gross Profit Margin'] = (rating_df['Gross Profit (Loss)'] / rating_df['Total Revenue']) * 100
    print("Successfully added 'Gross Profit Margin'.")
    
else:
    print("Required columns 'Gross Profit (Loss)' or 'Total Revenue' are missing.")


rating_df = rating_df.replace([float('inf'), float('-inf')], None)
rating_df

```

```python
#Step 12: Column Filter Node - remove unnecessary columns (axis=1 removes cols)
rating_df = rating_df.drop(
    ['Rating',
     'Total Current Liabilities', 
     'Non current asset'], 
    axis=1, 
    # Prevent errors if columns are already missing
    errors='ignore'  
)

rating_df
```

```python
#Step 13: Missing Value Node - drop rows that consist of missing values
rating_df.isnull().sum()
rating_df = rating_df.dropna()
rating_df


# Replace 'inf' and '-inf' with NaN
#rating_df.replace([float('inf'), float('-inf')], float('nan'), inplace=True)

# Drop rows with missing values (NaN)
#rating_df = rating_df.dropna().copy()

# Verify the new shape
#print(f"DataFrame shape after dropping 'inf' and NaN values: {rating_df.shape}")

# Check if the shape matches the expected result
#if rating_df.shape == (2586, 28):
#    print("The DataFrame now has 2586 rows and 28 columns as expected.")
#else:
#    print(f"Unexpected shape: {rating_df.shape}. Please recheck earlier steps.")


```
